```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
import pandas as pd

# Load data from file
a = np.loadtxt('PL_RMSD.dat')
x = a[:, 0]
y = a[:, 5]

# Perform rolling smoothing on the data
smoothed_y = pd.Series(y).rolling(window=10, center=True).mean()

# Create the plot
fig, ax = plt.subplots()
line, = ax.plot(x, smoothed_y)
dot, = ax.plot([], [], 'ro')

# Set plot properties
ax.set_xlabel('Time (ns)',fontsize =15)
ax.set_ylabel('Ligand RMSD ()', fontsize =15)
#ax.set_title('Ligand RMSD vs. Time with dye on')
ax.set_xlim(0,1000)

# Set the initial position of the dot
dot.set_data([], [])

# Select frames with stride
stride = 10
frames = np.arange(0, len(x), stride)

# Update function for the animation
def update(frame):
    dot.set_data(x[frame], smoothed_y[frame])
    return dot,

# Create the animation
ani = animation.FuncAnimation(fig, update, frames=frames, interval=160, blit=True)

# Save the animation as an MP4 file
ani.save('dye_on_animated_plot2.mp4', writer='ffmpeg', dpi=300)

# Increase tick size
ax.tick_params(axis='both', which='major', labelsize=12)
# Show the initial plot
plt.show()










```
